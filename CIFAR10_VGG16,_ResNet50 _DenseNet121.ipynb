CIFAR10 DATASET

Classify CIFAR10 dataset with pretrained models - VGG16, ResNet50, DenseNet121 and analyze the performance

```python
#import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.densenet import DenseNet121
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
# Load the CIFAR10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# Normalize pixel values between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
# Convert labels to one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
# Load the VGG16 model and add additional layers
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
model_vgg = Sequential()
model_vgg.add(vgg_model)
model_vgg.add(Flatten())
model_vgg.add(Dense(256, activation='relu'))
model_vgg.add(Dense(10, activation='softmax'))
```

```python
# Load the ResNet50 model and add additional layers
resnet_model = ResNet50(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
model_resnet = Sequential()
model_resnet.add(resnet_model)
model_resnet.add(Flatten())
model_resnet.add(Dense(256, activation='relu'))
model_resnet.add(Dense(10, activation='softmax'))
```

```python
# Load the DenseNet121 model and add additional layers
densenet_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
model_densenet = Sequential()
model_densenet.add(densenet_model)
model_densenet.add(Flatten())
model_densenet.add(Dense(256, activation='relu'))
model_densenet.add(Dense(10, activation='softmax'))
```

```python
# Compile and train the VGG16 model
model_vgg.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history_vgg = model_vgg.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))
```

```python
model_vgg.summary()
```

```python
# Compile and train the ResNet50 model
model_resnet.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history_resnet = model_resnet.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))
```

```python
model_resnet.summary()
```

```python
# Compile and train the DenseNet121 model
model_densenet.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history_densenet = model_densenet.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))
```

```python
model_densenet.summary()
```

```python
# Evaluate the VGG16 model
loss_vgg, acc_vgg = model_vgg.evaluate(x_test, y_test, verbose=0)
print(loss_vgg)
print(acc_vgg)
```

```python
# Evaluate the resnet model
loss_resnet, acc_resnet = model_resnet.evaluate(x_test, y_test, verbose=0)
print(loss_resnet)
print(acc_resnet)
```

```python
# Evaluate the ResNet model
loss_densenet, acc_densenet = model_densenet.evaluate(x_test, y_test, verbose=0)
print(loss_densenet)
print(acc_densenet)
```
